```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = r"C:.txt" #.txt or .csv file path
df = pd.read_csv(file_path, sep="\t", header=0)

x = df[f'Re(Z)/Ohm']
y = df[f'-Im(Z)/Ohm']
plt.scatter(x,y, s=10)
plt.xlim(-5, 350)
plt.ylim(-5, 350)
plt.title("Nyquist Plot for NFMMO_800C (Post 100 Cycles)")
plt.show()

```
