## Capacity Fade

```python
from battery_characterisation_tool.echem_plotting.capacity_fade_plotting import CapacityFadePlotting
folder_path = rf"C:" #this time navigate to the folder path not a specific data set. The data will be plotted in the order that the files are listed in the folder.
dataset_name = "Muon P2 Samples"
plot_title = "Capacity Fade Plot"
active_mass_list = [0.01230, 0.01114, 0.01016] #add in as many active masses in g as you wish to plot on one graph
xlim = (-2, 60)
legend_labels = ["P2", "P2Ca", "P2CaSr"] #have the same number of labels as number of active masses above.
plot = CapacityFadePlotting(folder_path = folder_path,
                            dataset_name = dataset_name,
                            plot_title = plot_title,
                            active_mass_list = active_mass_list,
                            xlim = xlim,
                            legend_labels=legend_labels)
plot.capacity_fade()
```
